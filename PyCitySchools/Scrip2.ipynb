```python
#Total number of unique schools
unique_Schools=len(pd.unique(schools_df['school_name']))
unique_Schools

print(f'There are '+ str(unique_Schools)+ ' schools')
```

```python
#Total students
total_students=students_df["student_name"].count()
print(f'The total of students is : '+ str(total_students))
```

```python
#Total budget
total_budget=schools_df.loc[:,'budget'].sum()
print(f'The Total of the budget is :  {total_budget}')
```

```python
#Average reading score
reading_score_aver=students_df.loc[:,'reading_score'].mean().__round__(2)
print(f'The average of the reading score is :  {reading_score_aver} %')

#Average math score
math_score_aver=students_df.loc[:,'math_score'].mean().__round__(2)
print(f'The average of the math score is :  {math_score_aver} %')
```
